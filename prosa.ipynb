```python
import pandas as pd
data=pd.read_csv(r"C:\Users\sonusnair\Downloads\kuttu ma kuttu.csv")
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
x = data[['CO','Ozone','NO','NO2','NOX','NH3','SO2','PM2.5','PM10']]
y = data['AQI']
y
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train
y
```

```python

from sklearn.linear_model import LinearRegression
li = LinearRegression()
li.fit(x_train,y_train)
```

```python

pre=li.predict(x_test)
pre
```

```python
import pickle
f=open('li.pkl', 'wb')
pickle.dump(li, f)
f.close()
```

```python

f=open('li.pkl','rb')
li=pickle.load(f)
f.close()
```

```python

from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train, y_train)
```

```python

data=dt_regressor.predict(x_test)
data
```

```python

from sklearn.metrics import r2_score
y_pred = li.predict(x_test)
r2_score(y_test, y_pred)
```

```python

from sklearn.metrics import mean_squared_error
print("MSE", mean_squared_error(y_test, y_pred))
```

```python
import numpy as np
print("RME",np.sqrt(mean_squared_error(y_test, y_pred)))
```
